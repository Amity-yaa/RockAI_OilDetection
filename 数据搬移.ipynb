### 本数据搬移是希望将读取的图片缩放减小，加快数据读取速度

```python
import os 
import shutil

orig_dir = '原始数据'
if not os.path.exists(orig_dir):
    os.mkdir(orig_dir)
train_1_root = 'train_1'
train_2_root = 'train_2'
fs_1 = os.listdir(train_1_root)
fs_2 = os.listdir(train_2_root)
```

### 随机抽10张train_1, train_2的图片观察图片大小是否是正方形，虽然是不是正方形，只要差距不是离谱都没有问题，但是好奇

```python
import numpy as np 
from PIL import Image
for i in range(10):
    filename = np.random.choice(fs_1,1)[0]
    filepath = os.path.join(train_1_root,filename)
    print(filepath,Image.open(filepath).size)
    
for i in range(10):
    filename = np.random.choice(fs_1,1)[0]
    filepath = os.path.join(train_1_root,filename)
    print(filepath,Image.open(filepath).size)
```

```python
train_1_new = os.path.join(orig_dir,train_1_root)
train_2_new = os.path.join(orig_dir,train_2_root)
if not os.path.exists(train_1_new):
    os.mkdir(train_1_new)
if not os.path.exists(train_2_new):
    os.mkdir(train_2_new)
    
for file in fs_1:
    file_path = os.path.join(train_1_root,file)
    file_new_path = file_path.replace(train_1_root,orig_dir+'/'+train_1_root)
    if not os.path.exists(file_new_path):
        shutil.copy(file_path,file_new_path)
    Image.open(file_path).convert('RGB').resize((224,224)).save(file_path)
    print(file_path,' 已resize')

# train_2文件夹不是做深度学习的，搞错了，不用做resize
# for file in fs_2:
#     file_path = os.path.join(train_2_root,file)
#     file_new_path = file_path.replace(train_2_root,orig_dir+'/'+train_2_root)
#     shutil.copy(file_path,file_new_path)
#     Image.open(file_path).convert('RGB').resize((224,224)).save(file_path)
#     print(file_path,' 已resize')
```
